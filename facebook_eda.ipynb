```python
#Convert tenure to int64 type from float
fb_data['tenure']=fb_data['tenure'].astype('int64')
#Convert gender column object type to category column type 
fb_data['gender']= fb_data['gender'].astype('category')
```

```python
#use info method to learn the shape of object types of given data
fb_data.info()
```

```python
#checking the null value in the given data
fb_data.isnull().sum()
```

**After  checking the null value its seen that in gender as well as in tenure column there is some missing values.
In gender column there are 175 null values and in tenure column there are 2 null values.
In python there are several ways to fill up the null values so we are using mode/median for better accuracy in result and larger data set**

**Imputation of missing values**

**1 a) Replace the null values (NA) of gender column with mode**

```python
fb_data['gender'] = fb_data['gender'].fillna(fb_data['gender'].mode()[0])
```

```python
fb_data.isnull().sum()
```

**Imputation of missing values**

**b) Replace the null values (NA) of tenure column with median**

```python
fb_data['tenure'] = fb_data['tenure'].fillna(fb_data['tenure'].median())
```

```python
fb_data.isnull().sum()
```

**A .corr() method will be added to the data and passed as the first argument.**

**For an even easier interpretation, an argument called annot=True should be passed**

**In below heat map adjusting the axis,asthestics, changing the matrix shape
vmin which is the minimun value of the bar; vmax which is the maximum value of the bar, centre changing none to zero.
cmap argument used for specifying the colour
line width is boarder thickness; line color is colour of line; square argument is used for specifying the shape.**

**Reduce the number displayed and improve readability is to pass the argument to fmt='.3g'**

```python
size
sns.heatmap(fb_data.corr(),annot = True, fmt='.3g', vmin = -1, vmax = 1, center= 0,  cmap = 'BrBG', linewidths = 3, linecolor= 'black', square= True)
```

**5) Analysis based on the user accessibility (Mobile Devices vs. Web Devices)**

**a) What is the average number of posts liked by users (based on gender) through web vs.
mobile devices?**

```python
size
scatter_plot = pd.concat([fb_data['mobile_likes'],fb_data['www_likes']],axis = 1)
sns.regplot(x='mobile_likes',y = 'www_likes',data = scatter_plot,scatter= True, fit_reg=True)
```

**b) What is the average number of likes received by users (based on gender) through web
vs. mobile devices?**

```python
size
scatter_plot = pd.concat([fb_data['mobile_likes_received'],fb_data['www_likes']],axis = 1)
sns.regplot(x='mobile_likes_received',y = 'www_likes',data = scatter_plot,scatter= True, fit_reg=True)
```
